## 10.4 Triangular Matrix


```python
M = np.array([
             [1, 2, 3],
             [1, 2, 3],
             [1, 2, 3]
             ])
print("M : \n", M)
print("lower M : \n", np.tril(M))
print("upper M : \n", np.triu(M))
```

## 10.5 Diagonal Matrix


```python
M = np.array([
             [8, 2, 3],
             [1, 9, 3],
             [1, 2, 10]
             ])
print("M : \n", M)

d = np.diag(M)
print("exact diagonal : \n", d)
print("create diagonal matrix from vector (ex 1): \n", np.diag(d))
print("create diagonal matrix from vector (ex 2): \n", np.diag(np.diag(M)))
```

## 10.6 Identity Matrix


```python
I = np.identity(3)
print("I : ", I)

I = np.identity(2)
print("I : ", I)
```

## 10.7 Orthogonal Matrix


```python
Q = np.array([
    [1, 0],
    [0, -1]
    ])
print("Q : \n", Q)

# inverse equivalence
V = np.linalg.inv(Q)
print("inverse equivalence (ex 1) : \n", Q.T)
print("inverse equivalence (ex 2) : \n", np.transpose(Q))

print("inverse equivalence : \n", V)

# identity equivalence
I = np.dot(Q, Q.T)
print("\nidentity equivalence : \n", I)
```
