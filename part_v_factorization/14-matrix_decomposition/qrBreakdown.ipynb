```python
import numpy as np
```

# Formula 


```python
D = np.array([
             [1, 2],
             [3, 4],
             [5, 6]
             ])

a = D[ :, 0]
b = D[ :, 1]
# print("a : ", a, "\nb : ", b)

A = a
# print("A : ",A)

proj_a_b = np.dot(np.dot(np.transpose(A), b) / np.dot(np.transpose(A), A), A)
print("\nproj_a_b : ", proj_a_b)

B = b - proj_a_b
print("\nB : ", B)
```

```python
ele_of_A = np.sqrt(np.sum(np.square(a)))
print("ele_of_A : ", ele_of_A)

q1 = A / ele_of_A 
print("q1 : ", q1)

ele_of_B = np.sqrt(np.sum(np.square(B)))
print("\nele_of_B : ", ele_of_B)

q2 = B / ele_of_B
print("q2 : ", q2)
```

```python
new_Q = np.array([])
print("new_Q : ", new_Q)

n_Q = np.vstack((q1, q2))
print("\nn_Q.shape : ", n_Q.shape, "\nn_Q : ", n_Q)
```

```python
R = np.dot(n_Q, D)
print("R : ", R)
```

```python
# Chceck if A = Q^T dot R
```

```python
n_Q_T = np.transpose(n_Q)

A = np.dot(n_Q_T, R)
print("A : ", A)
```
