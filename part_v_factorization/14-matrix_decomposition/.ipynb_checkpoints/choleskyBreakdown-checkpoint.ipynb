# Breakdown of Cholesky Decomposition
[Ex](https://atozmath.com/example/MatrixEv.aspx?he=e&q=choleskydecomp)

```python
import numpy as np
```

# Formula


```python
L = np.array([
             [1.414, 0, 0],
             [0.707, 1.225, 0],
             [0.707, 0.408, 1.155]
])

L_T = np.transpose(L)
A = np.dot(L, L_T)
print("A : ", A)
```

```python
# round up
A_up = np.around(A)
print("A_up : ", A_up)
```
