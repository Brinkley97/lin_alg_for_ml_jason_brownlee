```python
import numpy as np
```

## det(A)


```python
A = np.array([
             [1, 0, 1],
             [0, 1, 1],
             [1, 1, 1]
             ])
det_A = np.linalg.det(A)
print("det_A : \n", det_A)
```

## Matrix of Minors


# det(minor)


# det(A)$^($$ ^r $$ ^o $$ ^w $ $ ^+ $ $ ^c $$ ^o $$ ^l $$^)$


```python
A_minor = np.array([
             [0, 1, -1],
             [1, 0, -1],
             [-1, -1, 1]
             ])

print("A_minor : \n", A_minor)
```

# $ A^{-1} $



```python
one_over_det_A = 1 / det_A
A_minor_T = np.transpose(A_minor)

A_inv = np.dot(one_over_det_A, A_minor_T)
print("A_inv : \n", A_inv)
```
