```python
import numpy as np
```

```python
## 5 x 1
d4 = np.array([
              [8], 
              [0], 
              [0], 
              [0], 
              [0]
              ])

print("\n4 : ", d4)

dr4 = np.linalg.matrix_rank(d4)
print("dr4 : ", dr4)


## 5 x 2
d5 = np.array([
              [8, 0], 
              [0, 0], 
              [0, 0], 
              [0, 0], 
              [0, 0]
              ])

print("\n5 : ", d5)

dr5 = np.linalg.matrix_rank(d5)
print("dr5 : ", dr5)

## 4 x 5
d6 = np.array([
              [1, 0, -1, 0, 4], 
              [2, 1, 0, 0, 9], 
              [-1, 2, 5, 1, -5], 
              [1, -1, -3, 2, 9], 
              ])

print("\n6 : \n", d6)

dr6 = np.linalg.matrix_rank(d6)
print("dr6 : ", dr6)

## 2 x 2
d7 = np.array([
              [2, -3], 
              [-6, 9]
              ])

print("\n7 : \n", d7)

dr7 = np.linalg.matrix_rank(d7)
print("dr7 : ", dr7)


## 3 x 4
d8 = np.array([
              [2, -5, 5, 17], 
              [-1, 3, 0, -4],
              [1, -2, 3, 9]
              ])

print("\n8 : \n", d8)

dr8 = np.linalg.matrix_rank(d8)
print("dr8 : ", dr8)

d9 = np.array([
              [1, -2, 3, 9], 
              [-1, 3, 0, -4],
              [2, -5, 5, 17]
              ])

print("\n9 : \n", d9)

dr9 = np.linalg.matrix_rank(d9)
print("dr9 : ", dr9)

d10 = np.array([
              [3, 1, 0, -1], 
              [2, 4, 3, 2]
              ])

print("\n10 : \n", d10)

dr10 = np.linalg.matrix_rank(d10)
print("dr10 : ", dr10)

d11 = np.array([
              [3, 2, 4], 
              [1, -2, 3],
              [-3, -10, 1]
              ])

print("\n11 : \n", d11)

dr11 = np.linalg.matrix_rank(d11)
print("dr11 : ", dr11)

d12 = np.array([
              [5, 2, 43], 
              [7, 2, 2],
              [9, -1, 1]
              ])

print("\n12 : \n", d12)

dr12 = np.linalg.matrix_rank(d12)
print("dr12 : ", dr12)
```
