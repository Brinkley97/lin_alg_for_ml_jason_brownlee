## 7.4 Vector Arithmetic
1. All operations are element-wise between two vecs of same length
2. Result in a new vec of same len

### 7.4.1 Vector Addition

#### 1. How to compute the Vec Addition?
1. Notation : d = b + j
    1. b : b$ _1 $$ _i $
    2. j : j$ _1 $$ _i $
    3. d : d$ _1 $$ _i $
    4. $ \therefore $ all vecs must be the same size/ len
*** 


```python
a13 = np.array([1, 2, 3])
b13 = np.array([4, 5, 6])

c = a13 + b13
print("c : ", c)

# don't work bc diff sizes
a14 = np.array([1, 2, 3, 5])
b13 = np.array([4, 5, 6])

c = a14 + b13
print("c : ", c)
```

### 7.4.2 Vector Subtraction


```python
a13 = np.array([1, 2, 3])
b13 = np.array([4, 5, 6])

c = a13 - b13
print("c : ", c)

# don't work bc diff sizes
a13 = np.array([1, 2, 3])
b14 = np.array([4, 5, 6, 7])

c = a13 - b14
print("c : ", c)
```

## My code for the Dot Product

```python
a13 = np.array([1, 2, 3])
b13 = np.array([1, 2, 3])

# new arr to store new values
c = []

# loop len of full arr & store each index as dp
for dp in range(len(a13)) :
    
#     get element of each arr & mult them
    dot = a13[dp] * b13[dp]
    
#     take output & add to c list
    c.append(dot)

# total up elements in c list
c = np.sum(c)

print("c : ", c)
```
